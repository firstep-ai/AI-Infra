```python
# pip install datasets
import torch
from datasets import load_dataset
from transformers import AutoProcessor, AutoModelForVision2Seq

processor = AutoProcessor.from_pretrained("ByteDance/Dolphin")
model = AutoModelForVision2Seq.from_pretrained("ByteDance/Dolphin")

dataset = load_dataset("hf-internal-testing/example-documents", split="test")
image = dataset[0]["image"]
task_prompt = f"<s>Parse the reading order of this document. <Answer/>"
inputs = processor(image, task_prompt, return_tensors="pt")

outputs = model.generate(
    input_ids=inputs.input_ids,
    pixel_values=inputs.pixel_values,
    max_length=512
)
answer = processor.decode(outputs[0][len(inputs['input_ids'][0]):], skip_special_tokens=True)
print(answer)
```

```python
from transformers import DonutProcessor, VisionEncoderDecoderModel
from PIL import Image
EXAMPLE_IMAGE_PATH = "para_1.jpg"
processor = DonutProcessor.from_pretrained("ByteDance/Dolphin")
model = VisionEncoderDecoderModel.from_pretrained("ByteDance/Dolphin")

task_prompt = f"<s>Parse the table in the image. <Answer/>"
image = [Image.open(EXAMPLE_IMAGE_PATH).convert('RGB')]
inputs = processor(image, task_prompt, return_tensors="pt")

outputs = model.generate(
    input_ids=inputs.input_ids,
    pixel_values=inputs.pixel_values,
    max_length=512
)
answer = processor.decode(outputs[0][len(inputs['input_ids'][0]):], skip_special_tokens=True)
print(answer)
```
